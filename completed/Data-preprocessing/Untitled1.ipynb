```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Data.csv')
```

```python
df
```

## Fill in missing values - Impute

```python
from sklearn.preprocessing import Imputer
```

```python
imputer = Imputer(missing_values='NaN', strategy='mean')
```

```python
df[['Age','Salary']]=imputer.fit_transform(df[['Age','Salary']])
```

```python
df
```

## Encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labele = LabelEncoder()
```

```python
df['Purchased']=labele.fit_transform(df['Purchased'])
```

```python
df
```

```python
labelc = LabelEncoder()
```

```python
df['Country'] = labelc.fit_transform(df['Country'])
```

```python
df
```

## Remove number bias of categorical values

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(categorical_features=[0])
```

```python
ohe.fit_transform(df).toarray()
```

```python
df.columns
```

```python
pd.DataFrame(ohe.fit_transform(df).toarray(), columns=['France', 'Germany', 'Spain','Age', 'Salary', 'Purchased'])
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```
