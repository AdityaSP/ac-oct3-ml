```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
df1 = pd.read_csv('user_ratings.data', sep='\t', names=column_names)
df1.head()
```

```python
movie_titles = pd.read_csv("id_titles")
movie_titles.head()
```

- merge on item_id using pd.merge() on item_id

```python
df = pd.merge( movie_titles,df1 ,on='item_id')
```

```python
df.head()
```

- ratings dataframe by taking a groupby on ratings and mean()

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.sort_values('rating', ascending=False).head()
```

```python
ratings['count'] = df.groupby('title')['rating'].count()
```

```python
ratings.head()
```

```python
ratings.sort_values('count', ascending=False)
```

```python
df.head()
```

```python
movie_mat = df.pivot_table(index='user_id', columns='title', values='rating')
```

```python
movie_mat.head(3)
```

```python
pf = movie_mat['Pulp Fiction (1994)']
ll = movie_mat['Liar Liar (1997)']
```

- use corrwith on the matrix and pass the data of the movie

```python
pf_corr = pd.DataFrame(movie_mat.corrwith(pf), columns=['corr'])
```

```python
pf_corr.sort_values('corr', ascending=False).head()
```

```python
ratings.head()
```

```python
pf_corr_rat = pf_corr.join(ratings,on='title')
```

```python
pf_corr_rat[pf_corr_rat['count']>100].sort_values('corr', ascending=False)[['corr','count']]
```

```python
def recommend(movie_name):
    mov = movie_mat[movie_name]
    movie_corr = pd.DataFrame(movie_mat.corrwith(mov), columns=['corr'])
    movie_corr_rat = movie_corr.join(ratings,on='title')
    return movie_corr_rat[movie_corr_rat['count']>150].sort_values('corr', ascending=False)[['corr','count']].head(10)
    
```

```python
recommend('Star Wars (1977)')
```
